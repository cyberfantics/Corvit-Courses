```python
# Function to recursively print tag names
def print_tag_names(element, level=0):
    if element.name is not None:  # Check if the tag has a name (i.e., it is an element, not a navigable string)
        print('  ' * level + element.name)
        for child in element.children:
            if hasattr(child, 'name'):  # Check if the child is an element (not a navigable string or comment)
                print_tag_names(child, level + 1)

# Start printing from the root of the document
t_name = print_tag_names(p_soup)
```

### Task 18

### Write a python program to print the elements that have specified string of a webpage

```python
def find_elements_with_string(url, search_string):
    
    # Parse the content of the page with BeautifulSoup
    soup = parseLink(url)

    # Find all elements in the page
    elements = soup.find_all(string=lambda text: search_string.lower() in text.lower())

    # Print the content of the found elements
    if elements:
        print(f"Elements containing the string '{search_string}':\n")
        for element in elements:
            print(element.strip())
            print("-" * 80)
    else:
        print(f"No elements found containing the string '{search_string}'.")
```

```python
find_elements_with_string('https://python.org', "Python")
```

### Task 19

### Write a python program to print content of elements that have specified id of web page

```python
def print_element_by_id(url, element_id):
    try:
        # Send a GET request to the specified URL
        soup = parseLink(url)

        # Find the element with the specified ID
        element = soup.find(id=element_id)

        # Print the content of the found element
        if element:
            print(f"Content of element with ID '{element_id}':\n")
            print(element.get_text(strip=True))
        else:
            print(f"No element found with ID '{element_id}'.")
    
    except requests.exceptions.RequestException as e:
        print(f"Error fetching the URL: {e}")
```

### Task 22

### Write a python program to find tag beneath other tags in a given html documents

```python
soup = parseLink(w_link)

def find_tags_beneath(soup, parent_tag_name, child_tag_name):
    """
    Find all occurrences of `child_tag_name` tags that are beneath `parent_tag_name` tags.
    
    :param soup: BeautifulSoup object
    :param parent_tag_name: Tag name to look for parent tags
    :param child_tag_name: Tag name to look for child tags beneath the parent tags
    :return: List of BeautifulSoup tags
    """
    parent_tags = soup.find_all(parent_tag_name)
    child_tags = []

    for parent in parent_tags:
        children = parent.find_all(child_tag_name, recursive=False)
        child_tags.extend(children)

    return child_tags
```

```python
# Find all <p> tags beneath <div> tags
parent_tag_name = 'div'
child_tag_name = 'p'
found_tags = find_tags_beneath(soup, parent_tag_name, child_tag_name)

# Print the results
print(f"Tags <{child_tag_name}> found beneath <{parent_tag_name}> tags:")
for tag in found_tags:
    print(tag.prettify())
```

```python
def find_direct_children(soup, parent_tag_name, child_tag_name):
    """
    Find all occurrences of `child_tag_name` tags that are direct children of `parent_tag_name` tags.
    
    :param soup: BeautifulSoup object
    :param parent_tag_name: Tag name to look for parent tags
    :param child_tag_name: Tag name to look for direct child tags beneath the parent tags
    :return: List of BeautifulSoup tags
    """
    parent_tags = soup.find_all(parent_tag_name)
    child_tags = []

    for parent in parent_tags:
        children = parent.find_all(child_tag_name, recursive=False)
        child_tags.extend(children)

    return child_tags
```

### Task 24

### Write a python program to find siblings of tags in a given html document

```python
def find_siblings(soup, tag_name, attribute_name=None, attribute_value=None):
    """
    Find siblings of the first occurrence of a tag with the specified attribute.

    :param soup: BeautifulSoup object
    :param tag_name: Name of the tag to find siblings for
    :param attribute_name: Name of the attribute to match (default is None, meaning no attribute filtering)
    :param attribute_value: Value of the attribute to match (default is None, meaning no attribute filtering)
    :return: List of BeautifulSoup tags representing siblings
    """
    # Find the first occurrence of the tag with the given attribute value
    if attribute_name is not None and attribute_value is not None:
        target_tag = soup.find(tag_name, {attribute_name: attribute_value})
    else:
        target_tag = soup.find(tag_name)
    
    if not target_tag:
        return []  # No tag found, return an empty list

    # Find all siblings (both next and previous)
    siblings = target_tag.find_all_next_sibling() + target_tag.find_all_previous_sibling()
    
    # Filter out None and remove duplicate tags if needed
    siblings = list(filter(None, siblings))
    
    return siblings
```

```python
def find_tags_by_class(soup, class_name):
    """
    Find all tags with the specified CSS class in the HTML document.
    
    :param soup: BeautifulSoup object
    :param class_name: CSS class name to search for
    :return: List of BeautifulSoup tags with the specified class
    """
    # Use find_all with class_ to search by CSS class
    tags = soup.find_all(class_=class_name)
    return tags
```

```python
def add_content_to_tag(soup, tag_name, attribute_name=None, attribute_value=None, content_to_add=""):
    """
    Add content to a specified tag in the HTML document.

    :param soup: BeautifulSoup object
    :param tag_name: Name of the tag to find
    :param attribute_name: Attribute name to filter by (default is None)
    :param attribute_value: Attribute value to filter by (default is None)
    :param content_to_add: Content to add inside the tag
    :return: None
    """
    # Find the tag with the specified attribute if provided
    if attribute_name and attribute_value:
        tag = soup.find(tag_name, {attribute_name: attribute_value})
    else:
        tag = soup.find(tag_name)

    if tag:
        # Append the new content to the tag
        tag.append(content_to_add)
```

### Task 28

### Write a python program to insert a new text within a url in a specified position

```python
def insert_text_in_url(soup, tag_name, attribute_name, attribute_value, insert_text, position):
    """
    Insert text within a URL at a specified position.

    :param soup: BeautifulSoup object
    :param tag_name: Name of the tag containing the URL
    :param attribute_name: Attribute name that holds the URL (e.g., 'href')
    :param attribute_value: Value of the attribute to match (e.g., URL to find)
    :param insert_text: Text to insert into the URL
    :param position: Position in the URL to insert the text
    :return: None
    """
    # Find the tag with the specified attribute value
    tag = soup.find(tag_name, {attribute_name: attribute_value})

    if tag and attribute_name in tag.attrs:
        # Get the current URL
        url = tag[attribute_name]
        # Insert text at the specified position
        new_url = url[:position] + insert_text + url[position:]
        # Update the tag's attribute with the new URL
        tag[attribute_name] = new_url


# Insert text into the URL in the <a> tag with a specific href
insert_text_in_url(soup, 'link', 'href', '/static/bootstrap.min.css', '/new-path', 8)

# Print the modified HTML
print(soup.prettify())
```

```python
def insert_before_tag(soup, target_tag_name, insert_content):
    """
    Insert a new tag or string immediately before the specified tag(s).

    :param soup: BeautifulSoup object
    :param target_tag_name: Name of the tag to insert content before
    :param insert_content: New content to insert (string or BeautifulSoup Tag)
    :return: None
    """
    # Find the first occurrence of the target tag
    target_tag = soup.find(target_tag_name)

    if target_tag:
        # Insert the new content before the target tag
        if isinstance(insert_content, str):
            target_tag.insert_before(insert_content)
        elif isinstance(insert_content, Tag):
            target_tag.insert_before(insert_content)
        else:
            raise ValueError("insert_content must be a string or BeautifulSoup Tag.")


# Create new content to insert
new_tag = soup.new_tag('div')
new_tag.string = "Inserted content"
new_string = "Inserted string content"

# Insert new tag and string before the first <p> tag
insert_before_tag(soup, 'p', new_tag)
insert_before_tag(soup, 'p', new_string)

# Print the modified HTML
print(soup.prettify())
```

### Task 31

### Write a python program to remove the contents of tag in a given html document

```python
def clear_tag_content(soup, tag_name, attribute_name=None, attribute_value=None):
    """
    Remove the contents of the specified tag(s) in the HTML document.

    :param soup: BeautifulSoup object
    :param tag_name: Name of the tag to clear content from
    :param attribute_name: Attribute name to filter by (default is None)
    :param attribute_value: Attribute value to filter by (default is None)
    :return: None
    """
    # Find the tags with the specified attribute if provided
    if attribute_name and attribute_value:
        tags = soup.find_all(tag_name, {attribute_name: attribute_value})
    else:
        tags = soup.find_all(tag_name)

    # Clear the content of each found tag
    for tag in tags:
        tag.clear()
```

### Task 32

### Write a python program to extract a tag or string from a given tree of html document

```python
def extract_tag_or_string(soup, tag_name=None, attribute_name=None, attribute_value=None, use_css_selector=False, css_selector=None):
    """
    Extract a tag or string from the HTML document.

    :param soup: BeautifulSoup object
    :param tag_name: Name of the tag to extract (default is None)
    :param attribute_name: Attribute name to filter by (default is None)
    :param attribute_value: Attribute value to filter by (default is None)
    :param use_css_selector: Whether to use CSS selector instead of tag name (default is False)
    :param css_selector: CSS selector to use (required if use_css_selector is True)
    :return: List of extracted tags or strings
    """
    extracted_elements = []

    if use_css_selector:
        if css_selector:
            extracted_elements = soup.select(css_selector)
        else:
            raise ValueError("CSS selector must be provided if use_css_selector is True.")
    else:
        if attribute_name and attribute_value:
            extracted_elements = soup.find_all(tag_name, {attribute_name: attribute_value})
        else:
            extracted_elements = soup.find_all(tag_name)

    return [element for element in extracted_elements]
```

### Task 33

### Write a python program to remove a tag from a given tree of html document and destroy its contents

```python
def remove_tag_and_contents(soup, tag_name, attribute_name=None, attribute_value=None):
    """
    Remove a tag from the HTML document and destroy its contents.

    :param soup: BeautifulSoup object
    :param tag_name: Name of the tag to remove
    :param attribute_name: Attribute name to filter by (default is None)
    :param attribute_value: Attribute value to filter by (default is None)
    :return: None
    """
    # Find the tags with the specified attribute if provided
    if attribute_name and attribute_value:
        tags = soup.find_all(tag_name, {attribute_name: attribute_value})
    else:
        tags = soup.find_all(tag_name)

    # Decompose each found tag to remove it and its contents
    for tag in tags:
        tag.decompose()

```

### Task 34

### Write a python program to remove a tag or string from a given tree of html document and replace it with given tag or string

```python
def replace_tag_or_string(soup, target_name=None, target_attribute=None, target_value=None, new_content=None, use_css_selector=False, css_selector=None):
    """
    Remove a tag or string from the HTML document and replace it with a given tag or string.

    :param soup: BeautifulSoup object
    :param target_name: Name of the tag to remove (default is None)
    :param target_attribute: Attribute name to filter by (default is None)
    :param target_value: Attribute value to filter by (default is None)
    :param new_content: New content to insert (string or BeautifulSoup Tag)
    :param use_css_selector: Whether to use CSS selector instead of tag name (default is False)
    :param css_selector: CSS selector to use (required if use_css_selector is True)
    :return: None
    """
    if use_css_selector:
        if css_selector:
            elements = soup.select(css_selector)
        else:
            raise ValueError("CSS selector must be provided if use_css_selector is True.")
    else:
        if target_attribute and target_value:
            elements = soup.find_all(target_name, {target_attribute: target_value})
        else:
            elements = soup.find_all(target_name)

    for element in elements:
        # Replace the old element with the new content
        if new_content:
            # Insert new content before removing the old element
            element.insert_before(new_content)
        # Remove the old element
        element.decompose()

```

### Task 35

### Write a python program to wrap an element in the specified tag and create the new wrapper

```python
def wrap_element(soup, target_tag_name, wrapper_tag_name, attribute_name=None, attribute_value=None):
    """
    Wrap an existing element in a new tag and create the new wrapper.

    :param soup: BeautifulSoup object
    :param target_tag_name: Name of the tag to be wrapped
    :param wrapper_tag_name: Name of the new tag to act as the wrapper
    :param attribute_name: Attribute name to filter by (default is None)
    :param attribute_value: Attribute value to filter by (default is None)
    :return: None
    """
    # Find the tags with the specified attribute if provided
    if attribute_name and attribute_value:
        target_tags = soup.find_all(target_tag_name, {attribute_name: attribute_value})
    else:
        target_tags = soup.find_all(target_tag_name)
    
    # Create the new wrapper tag
    wrapper_tag = soup.new_tag(wrapper_tag_name)
    
    for tag in target_tags:
        # Extract the contents of the target tag
        contents = tag.contents
        
        # Create a new wrapper tag if it's not already created
        new_wrapper = wrapper_tag.copy()
        
        # Insert the contents into the new wrapper tag
        for content in contents:
            new_wrapper.append(content)
        
        # Replace the target tag with the new wrapper
        tag.replace_with(new_wrapper)
```

### Task 36

### Write a python program to replace a given tag with whatever's inside a given tag

```python
def replace_tag_with_contents(soup, target_tag_name, source_tag_name):
    """
    Replace a given tag with the contents of another specified tag.

    :param soup: BeautifulSoup object
    :param target_tag_name: Name of the tag to be replaced
    :param source_tag_name: Name of the tag whose contents will be used for replacement
    :return: None
    """
    # Find the first occurrence of the source tag
    source_tag = soup.find(source_tag_name)
    
    if not source_tag:
        raise ValueError(f"No tag found with the name '{source_tag_name}'")

    # Get the contents of the source tag
    source_contents = source_tag.contents

    # Find all occurrences of the target tag
    target_tags = soup.find_all(target_tag_name)

    for target_tag in target_tags:
        # Replace the target tag with the contents of the source tag
        target_tag.replace_with(source_contents)

```
