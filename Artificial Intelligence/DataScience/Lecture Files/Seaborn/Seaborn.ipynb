# Upload Penguines Dataset

```python
df = pd.read_csv('penguins.csv')
```

```python
df
```

```python
sns.lineplot(x='bill_length_mm',y='flipper_length_mm',data=df)
plt.show()
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
sns.lineplot(x='bill_length_mm',y='flipper_length_mm',data=df)
plt.show()
```

```python
sns.lineplot(x='bill_length_mm',y='flipper_length_mm',hue='sex',data=df)
plt.show()
```

```python
sns.lineplot(x='bill_length_mm',y='flipper_length_mm',hue='species',data=df)
plt.show()
```

# Use own colors

```python
sns.lineplot(x='bill_length_mm',y='flipper_length_mm',data=df,hue='sex',palette='Spectral',style='sex',markers=['>','o'])
plt.show()
```

# Grid

```python
sns.lineplot(x='bill_length_mm',y='flipper_length_mm',data=df,hue='sex',style='sex',markers=['o','>'])
plt.title("This is Line Plot")
plt.grid()
plt.show()
```
