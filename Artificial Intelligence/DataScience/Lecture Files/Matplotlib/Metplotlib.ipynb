```python
import matplotlib.pyplot as plt
```

# Linear Graph

```python
x = [10,20,25,40,64]
y = [2,3,4,5,6]
```

```python
plt.plot(x,y)
plt.show()
```

# Change color of line

```python
c = 'r'
plt.plot(x,y,c)
plt.show()
```

# Marker

```python
x = [12,3,32,23,54]
y = [65,45,23,12,65]
```

```python
plt.plot(x,y,marker='o')
plt.show()
```

```python
plt.plot(x,y,marker='x')
plt.show()
```

# Marker on one axis

```python
plt.plot(x,y)
plt.plot(y,marker='<')
plt.plot(x,marker='o')
plt.show()
```

# Bar Graph

```python
x = ['A','B','C','D']
y = [34,67,45,34]
```

```python
plt.bar(x,y)
plt.show()
```

```python
c = ['r','b','g','y']
plt.bar(x,y,color =c)
plt.show()
```

### Horizontal Bar Graph

```python
plt.barh(x,y)
plt.show()
```

```python
plt.barh(x,y,color=c)
plt.show()
```

#### Having only one color

```python
plt.barh(x,y,color='red')
plt.show()
```

```python
plt.bar(x,y,color=c,width=0.3)
plt.show()
```

```python
plt.barh(x,y,color='red',height=0.2)
plt.show()
```

# Scatter Plot

```python
x = [1,2,4,5,6]
y = [45,34,65,78,54]
```

```python
plt.title('Linear Graph')
plt.plot(x,y)
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.title("Scater Plot")
plt.scatter(x,y,color='red')
plt.show()
```

#### Size in scatter plot

```python
c = ['red','green','pink','orange','blue']

plt.title("Scatter Plot")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.scatter(x,y,color=c,s=150)
plt.show()
```

```python
plt.scatter(x,y,color=c,s=200)
plt.show()
```

# Let's Play With Images

```python
import numpy as np
```

```python
from PIL import Image
```

```python
fname = 'image.jpg'
```

```python
image = Image.open(fname).convert('L')
```

### Mapping image using PIL

```python
plt.imshow(image,cmap='gray')
plt.show()
```

```python
plt.imshow(image,cmap='gray',alpha=0.5)
plt.show()
```

```python
plt.imshow(image,aspect=0.5,cmap='plasma')
plt.show()
```

```python
plt.title('OSINT Challenge Cat')
plt.xlabel("Reddit User's Cat")
plt.ylabel("Comments are Deleted")
plt.imshow(image,cmap='BrBG')
plt.show()
```
