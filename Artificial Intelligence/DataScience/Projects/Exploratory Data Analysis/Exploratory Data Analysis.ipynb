## Exploratory Data Anaysis
Exploratory data analysis is an approch of analyzing data sets to summarize their main characteristics, often using statistical graphics and other data visculization methods.

## Types of Exploratory Data Analysis
- Univariate Analysis
- Bivariate Analysis
- Multivariate Analysis
- Time Series Analysis
- Missing Data Analysis
- Outlier Analysis

```python
# import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read Dataset
df = pd.read_excel('US_Superstore_data.xls')
```

```python
# Shape of dataset
df.shape
```

```python
## Check head of dataset
df.head(1)
```

## Missing Value Analysis

```python
df.isnull().sum()
```

## Univarient Analysis

```python
df['Ship Mode'].value_counts()
```

```python
df['Category'].value_counts()
```

```python
df['Sub-Category'].value_counts()
```

```python
df['Category'].unique()
```

```python
plt.figure(figsize=(6,2))
sns.countplot(x='Ship Mode', data=df)
plt.title("Count of Ship Mode")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.countplot(df['Category'])
plt.title("Count of Title")
plt.show()
```

```python
plt.figure(figsize=(30,8))
sns.countplot(x=df['Sub-Category'])
plt.title("Count of Sub Category")
plt.show()
```

## Ship Mode

```python
## We check how much products are sell in which shiphing mode
df['Ship Mode'].value_counts()
```

```python
## Draw viscually
x = df['Ship Mode'].value_counts().index
y = df['Ship Mode'].value_counts().values
```

```python
plt.figure(figsize=(4,3))
plt.pie(y,labels=x, startangle=60, autopct="%0.2f%%")
plt.show()
```

## Product Category vs Ship Mode

```python
plt.figure(figsize=(5,3))
sns.countplot(x=df['Ship Mode'], hue=df['Category'])
plt.title("Analyze Ship Mode on the Base of Category")
plt.show()
```

```python
plt.figure(figsize=(17,3))
sns.countplot(x=df['Sub-Category'], hue=df['Category'])
plt.title("Analyze Sub Category on the base of Category")
plt.show()
```

## Supplies

```python
plt.figure(figsize=(5,3))
sns.countplot(x=df['Segment'])
plt.title('Analyze Coustomer Segment')
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.countplot(x=df['Segment'], hue=df['Category'])
plt.title("Check our Customer belong to which category Most of time")
plt.show()
```

## Country

```python
plt.figure(figsize=(4,3))
sns.countplot(x=df['Country']) ## It is unnecessary column in our dataset, as all of our coustomer belongs to only one country
plt.title("Our Coustomer belong to which Country")
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.countplot(x='Category', data=df[df['Category']=='Office Supplies'], hue=df['Sub-Category']) 
plt.show()
```

## Make a column for year

```python
df['Order Year'] = df['Order Date'].dt.year
```

```python
df.info()
```

```python
df['Order Year'].value_counts()
```

```python
plt.figure(figsize=(4,3))
sns.countplot(x=df['Order Year'])
plt.title("Years in which Orders Placed")
plt.show()
```

## PRofit Per Year

```python
plt.figure(figsize=(4,3))
sns.barplot(x=df['Order Year'], y = df['Profit'], estimator='sum')
plt.title("Profit on the Base of Year")
plt.show()
```
