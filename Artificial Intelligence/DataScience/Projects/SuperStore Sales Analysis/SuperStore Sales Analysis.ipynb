# `1. Time-Based Analysis`

#### Convert Order Date and Ship Date to datetime format

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

#### Calculate Time Difference

```python
# Calculate delivery time
df['Delivery Time'] = (df['Ship Date'] - df['Order Date']).dt.days
```

### Sales and Profit Trends Over Time

#### Sales Trend

```python
# Group by Order Date and aggregate sales
sales_trend = df.groupby('Order Date')['Sales'].sum()

# Plot sales trend
plt.figure(figsize=(12, 6))
sales_trend.plot()
plt.title('Sales Trend Over Time')
plt.xlabel('Order Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

### Profit Trend

```python
# Group by Order Date and aggregate profit
profit_trend = df.groupby('Order Date')['Profit'].sum()

# Plot profit trend
plt.figure(figsize=(12, 6))
profit_trend.plot()
plt.title('Profit Trend Over Time')
plt.xlabel('Order Date')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()
```

### Delivery Time Analysis

```python
# Plot delivery times distribution
plt.figure(figsize=(12, 6))
df['Delivery Time'].hist(bins=50)
plt.title('Distribution of Delivery Times')
plt.xlabel('Delivery Time (Days)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

###### We analyze that most of delivery take 4 to 5 days

# `Monthly Or Weekly Aggregate`

### Monthly Aggregate

```python
# Resample data to monthly frequency and aggregate sales
monthly_sales = df.resample('ME', on='Order Date')['Sales'].sum()

# Plot monthly sales
plt.figure(figsize=(12, 6))
monthly_sales.plot()
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

#### Weekly Aggregate

```python
# Resample data to weekly frequency and aggregate profit
weekly_profit = df.resample('W', on='Order Date')['Profit'].sum()

# Plot weekly profit
plt.figure(figsize=(12, 6))
weekly_profit.plot()
plt.title('Weekly Profit Trend')
plt.xlabel('Week')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()
```

### Seasoned Analysis

```python
# Add month and year columns for seasonal analysis
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month

# Group by Year and Month and aggregate sales
monthly_seasonal_sales = df.groupby(['Year', 'Month'])['Sales'].sum().unstack()

# Plot seasonal sales
plt.figure(figsize=(20, 10))
monthly_seasonal_sales.plot()
plt.title('Seasonal Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.legend(title='Year')
plt.grid(True)
plt.show()
```

# `2. Shipping and Logistics`

#### Distribution of Delivery time

```python
# Plot delivery times distribution with customized colors
plt.figure(figsize=(12, 6))
sns.histplot(df['Delivery Time'], bins=50, kde=True, color='royalblue')
plt.title('Distribution of Delivery Times')
plt.xlabel('Delivery Time (Days)')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

#### Shipping mode analysis
Analyze how different shipping modes impact delivery times and overall sales.

## Delivery Time by Shipping Mode

```python
# Group by Ship Mode and calculate average delivery time
ship_mode_delivery_time = df.groupby('Ship Mode')['Delivery Time'].mean()

# Plot delivery time by shipping mode
plt.figure(figsize=(12, 6))
ship_mode_delivery_time.plot(kind='bar', color='green')
plt.title('Average Delivery Time by Shipping Mode')
plt.xlabel('Shipping Mode')
plt.ylabel('Average Delivery Time (Days)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

### Sales and Profit by shipping mode

```python
# Group by Ship Mode and aggregate sales and profit
ship_mode_sales = df.groupby('Ship Mode')['Sales'].sum()
ship_mode_profit = df.groupby('Ship Mode')['Profit'].sum()

# Plot sales and profit by shipping mode
fig, ax = plt.subplots(2, 1, figsize=(12, 12))

# Sales plot
ship_mode_sales.plot(kind='bar', color='seagreen', ax=ax[0])
ax[0].set_title('Total Sales by Shipping Mode')
ax[0].set_xlabel('Shipping Mode')
ax[0].set_ylabel('Total Sales')
ax[0].grid(True, linestyle='--', alpha=0.7)

# Profit plot
ship_mode_profit.plot(kind='bar', color='blue', ax=ax[1])
ax[1].set_title('Total Profit by Shipping Mode')
ax[1].set_xlabel('Shipping Mode')
ax[1].set_ylabel('Total Profit')
ax[1].grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

### Geographical Analysis

#### Average Delivery Time by Region

```python
# Group by Region and calculate average delivery time
region_delivery_time = df.groupby('Region')['Delivery Time'].mean()

# Plot average delivery time by region
plt.figure(figsize=(12, 6))
region_delivery_time.plot(kind='bar', color='orchid')
plt.title('Average Delivery Time by Region')
plt.xlabel('Region')
plt.ylabel('Average Delivery Time (Days)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

### Total Sales and Profit by region

```python
# Group by Region and aggregate sales and profit
region_sales = df.groupby('Region')['Sales'].sum()
region_profit = df.groupby('Region')['Profit'].sum()

# Plot sales and profit by region
fig, ax = plt.subplots(2, 1, figsize=(12, 12))

# Sales plot
region_sales.plot(kind='bar', color='mediumslateblue', ax=ax[0])
ax[0].set_title('Total Sales by Region')
ax[0].set_xlabel('Region')
ax[0].set_ylabel('Total Sales')
ax[0].grid(True, linestyle='--', alpha=0.7)

# Profit plot
region_profit.plot(kind='bar', color='darkorange', ax=ax[1])
ax[1].set_title('Total Profit by Region')
ax[1].set_xlabel('Region')
ax[1].set_ylabel('Total Profit')
ax[1].grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

### Shipping Mode and Sales Correlation

```python
# Group by Ship Mode and calculate total sales and profit
ship_mode_data = df.groupby('Ship Mode').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()

# Plot sales vs. profit for different shipping modes
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Sales', y='Profit', hue='Ship Mode', data=ship_mode_data, palette='tab10', s=100)
plt.title('Sales vs. Profit by Shipping Mode')
plt.xlabel('Total Sales')
plt.ylabel('Total Profit')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

# `3. Coustomer Insight`

### Total Sales and Profit By Customer

```python
# Group by Customer ID and calculate total sales and profit
customer_sales_profit = df.groupby('Customer ID').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()

# Plot total sales and profit for each customer
fig, ax = plt.subplots(2, 1, figsize=(12, 12))

# Sales plot
sns.histplot(customer_sales_profit['Sales'], bins=50, kde=True, color='royalblue', ax=ax[0])
ax[0].set_title('Distribution of Total Sales per Customer')
ax[0].set_xlabel('Total Sales')
ax[0].set_ylabel('Frequency')
ax[0].grid(True, linestyle='--', alpha=0.7)

# Profit plot
sns.histplot(customer_sales_profit['Profit'], bins=50, kde=True, color='darkorange', ax=ax[1])
ax[1].set_title('Distribution of Total Profit per Customer')
ax[1].set_xlabel('Total Profit')
ax[1].set_ylabel('Frequency')
ax[1].grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

### Top Customers By Sales

```python
# Sort customers by total sales
top_customers = customer_sales_profit.sort_values(by='Sales', ascending=False).head(10)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Customer ID', y='Sales', data=top_customers, palette='viridis', hue=None)
plt.title('Top 10 Customers by Total Sales')
plt.xlabel('Customer ID')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

### Customer Behaviour Analysis

```python
# Group by Customer ID and Order Date, then count the number of orders
order_frequency = df.groupby('Customer ID').size().reset_index(name='Order Count')

# Plot purchase frequency
plt.figure(figsize=(12, 6))
sns.histplot(order_frequency['Order Count'], bins=30, kde=True, color='teal')
plt.title('Distribution of Purchase Frequency per Customer')
plt.xlabel('Number of Orders')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

#### Average Sales and Profit Per Order

```python
# Calculate average sales and profit per order
df['Order Count'] = df.groupby('Customer ID').cumcount() + 1
average_sales_profit_per_order = df.groupby('Customer ID').agg({'Sales': 'mean', 'Profit': 'mean'}).reset_index()

# Plot average sales and profit per order
fig, ax = plt.subplots(2, 1, figsize=(12, 12))

# Average Sales plot
sns.histplot(average_sales_profit_per_order['Sales'], bins=30, kde=True, color='royalblue', ax=ax[0])
ax[0].set_title('Distribution of Average Sales per Order per Customer')
ax[0].set_xlabel('Average Sales')
ax[0].set_ylabel('Frequency')
ax[0].grid(True, linestyle='--', alpha=0.7)

# Average Profit plot
sns.histplot(average_sales_profit_per_order['Profit'], bins=30, kde=True, color='darkorange', ax=ax[1])
ax[1].set_title('Distribution of Average Profit per Order per Customer')
ax[1].set_xlabel('Average Profit')
ax[1].set_ylabel('Frequency')
ax[1].grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

# `5. Performance Metrics`

```python
# Calculate average Sales, Quantity, and Profit
average_sales = df['Sales'].mean()
average_quantity = df['Quantity'].mean()
average_profit = df['Profit'].mean()

print(f"Average Sales: ${average_sales:.2f}")
print(f"Average Quantity: {average_quantity:.2f}")
print(f"Average Profit: ${average_profit:.2f}")
```

```python
# Convert Order Date to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Extract Year and Month
df['YearMonth'] = df['Order Date'].dt.to_period('M')

# Aggregate sales by YearMonth
monthly_sales = df.groupby('YearMonth').agg({'Sales': 'sum'}).reset_index()

# Calculate Sales Growth
monthly_sales['Sales Growth'] = monthly_sales['Sales'].pct_change() * 100
```

```python

# Plot Monthly Sales Growth
plt.figure(figsize=(14, 7))
sns.barplot(x='YearMonth', y='Sales Growth', data=monthly_sales, palette='viridis')
plt.title('Monthly Sales Growth')
plt.xlabel('Month')
plt.ylabel('Sales Growth (%)')
plt.xticks(rotation=90)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

## Annual Sales Growth

```python
# Extract Year
df['Year'] = df['Order Date'].dt.year

# Aggregate sales by Year
annual_sales = df.groupby('Year').agg({'Sales': 'sum'}).reset_index()

# Calculate Sales Growth
annual_sales['Sales Growth'] = annual_sales['Sales'].pct_change() * 100

```

```python

# Plot Annual Sales
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Sales', data=annual_sales, marker='o', color='b')
plt.title('Annual Sales')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# Plot Annual Sales Growth
plt.figure(figsize=(12, 6))
sns.barplot(x='Year', y='Sales Growth', data=annual_sales, palette='viridis')
plt.title('Annual Sales Growth')
plt.xlabel('Year')
plt.ylabel('Sales Growth (%)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
