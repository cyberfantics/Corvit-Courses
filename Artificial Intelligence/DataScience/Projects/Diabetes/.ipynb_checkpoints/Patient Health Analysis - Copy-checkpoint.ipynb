# `4. Data Cleaning and Manipulation`

#### Handle missing values

```python
fill_values = {
    'Age': df['Age'].median(),         # Median of 'Age'
    'Cholesterol': df['Cholesterol'].mean()  # Mean of 'Cholesterol'
}

df = df.fillna(value=fill_values)
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.columns
```

#### Organize Blood Pressure

```python
# Split the 'Blood Pressure (BP)' column into 'Systolic' and 'Diastolic'
df[['Systolic', 'Diastolic']] = df['Blood Pressure (BP)'].str.split('/', expand=True)

# Convert these new columns to numeric type
df['Systolic'] = pd.to_numeric(df['Systolic'], errors='coerce')
df['Diastolic'] = pd.to_numeric(df['Diastolic'], errors='coerce')

# Drop the original 'Blood Pressure (BP)' column
df.drop(columns=['Blood Pressure (BP)'], inplace=True)
```

```python
df.head(3)
```

```python
df.columns
```

#### Calculate New Metrics

```python
# Define BMI categories
def categorize_bmi(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 24.9:
        return 'Normal weight'
    elif 25 <= bmi < 29.9:
        return 'Overweight'
    else:
        return 'Obesity'
```

```python
df['BMI Category'] = df['Body Mass Index (BMI)'].apply(categorize_bmi)
```

#### Blood Pressure Categories

```python
def categorize_blood_pressure(systolic, diastolic):
    if systolic < 120 and diastolic < 80:
        return 'Normal'
    elif 120 <= systolic < 130 and diastolic < 80:
        return 'Elevated'
    elif 130 <= systolic < 140 or 80 <= diastolic < 90:
        return 'Hypertension Stage 1'
    else:
        return 'Hypertension Stage 2'
```

```python
# Apply the function to create a new column for Blood Pressure category
df['BP Category'] = df.apply(
    lambda row: categorize_blood_pressure(row['Systolic'], row['Diastolic']),
    axis=1
)
```

```python
df.head(1)
```

### Age Categories

```python
def age_categories(age):
    if age < 13:
        return 'Child'
    elif 13 <= age < 20:
        return 'Teenager'
    elif 20 <= age < 40:
        return 'Adult'
    elif 40 <= age < 60:
        return 'Middle-aged'
    else:
        return 'Senior'
```

```python
# Apply the age_categories function to create a new column for Age Category
df['Age Category'] = df['Age'].apply(age_categories)
```

### Cholestrol Categores

```python
def categorize_cholesterol(cholesterol):
    if cholesterol < 200:
        return 'Desirable'
    elif 200 <= cholesterol < 240:
        return 'Borderline High'
    else:
        return 'High'

# Apply the function to create a new column for Cholesterol Category
df['Cholesterol Category'] = df['Cholesterol Total'].apply(categorize_cholesterol)
```

# 5. Visculizing Blood Pressure Trends

```python
# Define the age categories
age_categories = df['Age Category'].unique()

# Plotting
plt.figure(figsize=(14, 6))

# Systolic Blood Pressure Box Plot
plt.subplot(1, 2, 1)
for category in age_categories:
    subset = df[df['Age Category'] == category]['Systolic']
    plt.boxplot(subset, positions=[list(age_categories).index(category)+1], labels=[category])
plt.title('Systolic Blood Pressure by Age Group')
plt.xlabel('Age Category')
plt.ylabel('Systolic Blood Pressure')

# Diastolic Blood Pressure Box Plot
plt.subplot(1, 2, 2)
for category in age_categories:
    subset = df[df['Age Category'] == category]['Diastolic']
    plt.boxplot(subset, positions=[list(age_categories).index(category)+1], labels=[category])
plt.title('Diastolic Blood Pressure by Age Group')
plt.xlabel('Age Category')
plt.ylabel('Diastolic Blood Pressure')

plt.tight_layout()
plt.show()
```

```python
# Function to create violin plots
def plot_violin(df, column, title):
    plt.figure(figsize=(12, 6))
    age_categories = df['Age Category'].unique()
    data = [df[df['Age Category'] == cat][column].dropna() for cat in age_categories]

    # Create violin plot
    plt.violinplot(data, showmeans=False, showmedians=True)

    plt.xticks(np.arange(1, len(age_categories) + 1), age_categories)
    plt.title(title)
    plt.xlabel('Age Category')
    plt.ylabel(column)

    plt.grid(True)
    plt.show()

# Plot Violin Plots
plot_violin(df, 'Systolic', 'Systolic Blood Pressure by Age Group')
plot_violin(df, 'Diastolic', 'Diastolic Blood Pressure by Age Group')
```

# 6. Cholesterol and Heart Rate Analysis:

```python
# Create scatter plot
plt.figure(figsize=(10, 6))

# Scatter plot for Cholesterol vs. Heart Rate
plt.scatter(df['Cholesterol Total'], df['Heart Rate'], color='blue',alpha=0.1, s=10)

# Add titles and labels
plt.title('Relationship Between Cholesterol Levels and Heart Rate')
plt.xlabel('Cholesterol (mg/dL)')
plt.ylabel('Heart Rate (bpm)')

# Show grid
plt.grid(True)

# Display the plot
plt.show()
```

```python
# Sample 1% of the data for visualization
sampled_df = df.sample(frac=0.01, random_state=1)  # Adjust fraction as needed

# Create scatter plot for sampled data
plt.figure(figsize=(12, 8))
plt.scatter(sampled_df['Cholesterol Total'], sampled_df['Heart Rate'], color='blue', alpha=0.7, s=10)
plt.title('Relationship Between Cholesterol Levels and Heart Rate (Sampled)')
plt.xlabel('Cholesterol (mg/dL)')
plt.ylabel('Heart Rate (bpm)')
plt.grid(True)
plt.show()
```

```python
# Scatter plot with categories
plt.figure(figsize=(12, 8))
categories = df['Cholesterol Category'].unique()

for category in categories:
    subset = df[df['Cholesterol Category'] == category]
    plt.scatter(subset['Cholesterol Total'], subset['Heart Rate'], label=category, alpha=0.7, s=10)

plt.title('Cholesterol Levels vs. Heart Rate by Cholesterol Category')
plt.xlabel('Cholesterol (mg/dL)')
plt.ylabel('Heart Rate (bpm)')
plt.legend(title='Cholesterol Category')
plt.grid(True)
plt.show()
```

# 7. Gender-Based Health Comparisons:

```python
# Calculate average metrics by gender
averages = df.groupby('Gender').agg({
    'Cholesterol Total': 'mean',
    'Body Mass Index (BMI)': 'mean',
    'Systolic': 'mean',
    'Diastolic': 'mean'
}).reset_index()
```

### Average Cholestrol

```python
# Plot average cholesterol
plt.figure(figsize=(10, 6))
plt.bar(averages['Gender'], averages['Cholesterol Total'], color=['blue', 'orange'])
plt.title('Average Cholesterol Levels by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Cholesterol (mg/dL)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

### Average Body Mass Index

```python
# Plot average BMI
plt.figure(figsize=(10, 6))
plt.bar(averages['Gender'], averages['Body Mass Index (BMI)'], color=['blue', 'pink'])
plt.title('Average Body Mass Index (BMI) by Gender')
plt.xlabel('Gender')
plt.ylabel('Average BMI')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
